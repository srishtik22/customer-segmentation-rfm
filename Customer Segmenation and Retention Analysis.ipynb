# RFM Anlysis

```python
#first convert invoice date into datetime dtype
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
```

```python
df.info()
```

```python
#set reference date
reference_date=df['InvoiceDate'].max()+ pd.Timedelta(days=1)
```

```python
reference_date
```

```python
#Create the RFM table
rfm=df.groupby('CustomerID').agg({'InvoiceDate': lambda x: (reference_date-x.max()).days,
                                  'InvoiceNo': 'nunique',
                                  'TotalAmount': 'sum' 
                                  })
```

```python
rfm
```

```python
rfm.columns=['Recency','Frequency','Monetary']
rfm
```

```python
rfm.shape #total customers are 4338
```

```python
rfm.isnull().sum()
```

```python
rfm.dtypes
```

```python
#Recency score
rfm['R_score']=pd.qcut(rfm['Recency'],5,[5,4,3,2,1])
rfm['F_score']=pd.qcut(rfm['Frequency'].rank(method='first'),5,[1,2,3,4,5])
rfm['M_score']=pd.qcut(rfm['Monetary'],5,[1,2,3,4,5])
```

```python
rfm
```

```python
# Create the combined RFM score
rfm['RFM_score']=rfm['R_score'].astype(int)+rfm['F_score'].astype(int)+rfm['M_score'].astype(int)
rfm
```

```python
#RFM Segment 
rfm['RFM_segment']=rfm['R_score'].astype(str)+rfm['F_score'].astype(str)+rfm['M_score'].astype(str)
```

# Segmentation

```python
def segmentation(score):
    if score>=13:
        return 'CHAMPION'
    elif score>=10:
        return 'LOYAL_CUSTOMERS'
    elif score>=6:
        return 'NEW_CUSTOMERS'
    elif score>=4:
        return 'AT_RISK' 
    else:
        return 'LOST_CUSTOMERS'

rfm['RFM_segmentation']=rfm['RFM_score'].apply(segmentation)
```

# Retention Analysis

```python
# How many customers purchased more than once?
rfm['repeat']=rfm['Frequency']>1
```

Frequency > 1  retained customer

Frequency = 1  one-time customer

```python
rfm['repeat'].value_counts()
```

True = retained

False = not retained

```python
#finding retention rate os each group by segment
rfm.groupby('RFM_segmentation')['repeat'].mean()*100
```

finding % of total spending by each segment group

```python
a=rfm.groupby('RFM_segmentation')['Monetary'].sum()
totala=a.sum()
totala
```

```python
b=(a/totala)*100
```

```python
b
```

```python
b.sum()
```

```python
# Which customer segments actually bring in the money?
# Revenue Contribution by RFM Segment (Percentage)

plt.figure(figsize=(8, 5))

plt.bar(b.index, b.values)

plt.xlabel('Customer Segmentation')
plt.ylabel('Revenue (%)')
plt.title('Revenue Contribution by RFM Segment (Percentage)')

plt.show()
```
